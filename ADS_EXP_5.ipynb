```python
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
from sklearn.datasets import make_classification
from imblearn.over_sampling import SMOTE
from imblearn.over_sampling import BorderlineSMOTE
```

```python
X, y = make_classification(n_samples=1000, n_features=2, n_redundant=0,
                           n_clusters_per_class=1, weights=[0.975], flip_y=0, random_state=1)
```

```python
counter=Counter(y)
counter
```

```python
for label, _ in counter.items():
    row_ix = np.where(y == label)[0]
    plt.scatter(X[row_ix, 0], X[row_ix, 1], label=str(label))
plt.title("Class Distribution before SMOTE")
plt.legend()
plt.show()
```

```python
oversample = SMOTE()
X1, y1 = oversample.fit_resample(X, y)

counter = Counter(y1)
counter
```

```python
for label, _ in counter.items():
    row_ix = np.where(y1 == label)[0]
    plt.scatter(X1[row_ix, 0], X1[row_ix, 1], label=str(label))
plt.title("Class Distribution after SMOTE")
plt.legend()
plt.show()
```

# BorderLine SMOTE

```python
oversample = BorderlineSMOTE()
X2, y2 = oversample.fit_resample(X, y)
# summarize the new class distribution
counter = Counter(y2)
print(counter)
```

```python
for label, _ in counter.items():
    row_ix = np.where(y2 == label)[0]
    plt.scatter(X2[row_ix, 0], X2[row_ix, 1], label=str(label))
plt.title("Class Distribution after Borderline SMOTE")
plt.legend()
plt.show()
```
